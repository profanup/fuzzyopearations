```python
from simpful import *
FS = FuzzySystem()

 # Define fuzzy sets and linguistic variables
I_1 = FuzzySet(function=Triangular_MF(a=0, b=0, c=50), term="small")
I_2 = FuzzySet(function=Triangular_MF(a=0, b=50, c=100), term="medium")
I_3 = FuzzySet(function=Triangular_MF(a=50, b=100, c=100), term="large")


FS.add_linguistic_variable("dirt", LinguisticVariable([I_1, I_2, I_3], concept="amount dirt", universe_of_discourse=[0,100]))

FS.add_linguistic_variable("grease", LinguisticVariable([I_1, I_2, I_3], concept="amount grease", universe_of_discourse=[0,100]))

# Define output fuzzy sets and linguistic variable
O_1 = FuzzySet(function=Triangular_MF(a=0, b=0, c=10), term="veryshort")
O_2 = FuzzySet(function=Triangular_MF(a=0, b=10, c=25), term="short")
O_3 = FuzzySet(function=Triangular_MF(a=10, b=25, c=40), term="medium")
O_4 = FuzzySet(function=Triangular_MF(a=25, b=40, c=60), term="long")
O_5 = FuzzySet(function=Triangular_MF(a=40, b=60, c=60), term="verylong")
FS.add_linguistic_variable("washtime", LinguisticVariable([O_1, O_2, O_3,O_4,O_5], universe_of_discourse=[0,60]))

# Define fuzzy rules
R1 = "IF (dirt IS small) AND (grease IS small) THEN (washtime IS veryshort)"
R2 = "IF (dirt IS small) AND (grease IS medium) THEN (washtime IS medium)"
R3 = "IF (dirt IS small) AND (grease IS large) THEN (washtime IS long)"
R4 = "IF (dirt IS medium) AND (grease IS small) THEN (washtime IS short)"
R5 = "IF (dirt IS medium) AND (grease IS medium) THEN (washtime IS medium)"
R6 = "IF (dirt IS medium) AND (grease IS large) THEN (washtime IS long)"
R7 = "IF (dirt IS large) AND (grease IS small) THEN (washtime IS medium)"
R8 = "IF (dirt IS large) AND (grease IS medium) THEN (washtime IS long)"
R9 = "IF (dirt IS large) AND (grease IS large) THEN (washtime IS verylong)"

FS.add_rules([R1, R2, R3, R4, R5,R6,R7,R8,R9])

# Set input values
FS.set_variable("dirt", 60)
FS.set_variable("grease", 60)

# Perform Mamdani inference and print output
print(FS.Mamdani_inference(["washtime"]))
```
