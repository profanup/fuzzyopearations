```python
'''union(max)'''
FA = dict()
FB = dict()
FC = dict()
FA={"a":0.5, "b":0.7, "c": 0.8}
FB={"a":0.7, "b": 0.8, "c":0.2}

for i in FA:
    a_value = FA[i]
    b_value = FB[i]
    if a_value > b_value:
      FC[i]= a_value
    else:
      FC[i]= b_value
print("Union:",FC)

'''intersection(min)'''
for i in FB:
    a_value = FA[i]
    b_value = FB[i]
    if a_value < b_value:
      FC[i]= a_value
    else:
      FC[i]= b_value
print("Intersection:",FC)

'''Negation'''
for i in FA:
   FC[i]=  1 - FA[i]
print("A:", FA)
print("Negation",FC)
```
